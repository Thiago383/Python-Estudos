## Declarando uma lista composta

Para declarar uma lista composta  s colocar uma lista dentro da outra

<h2> Entendendo como as listas compostas funcionam </h2>

Imagina que eu tenho um cardpio. Dentro dele eu tenho vrios comidas com seus respectivos preos.

O cardpio seria ento uma lista com todas as comidas

Cada comida vai ser uma lista com o Nome da Comida e o seu Preo.

```
comida1 = ['Lanche', 10.9]
comida2 = ['Batata',  5.5]
comida3 = ['Refri' ,  3.9]
cardapio = [comida1, comida2, comida3]
```
Ento acessando os termos temos:
```
cardapio[0]    == comida1 == ['Lanche', 10.9]
cardapio[0][1] == comida1[1] == 10.0
cardapio[2]    == comida3 == ['Refri' ,  3.9]
cardapio[2][0] == comida3[0] == 'Refri'
```
Vamos pegar uma lista dentro de outra lista j declarada e vamos entender como funciona acessar seus termos.

<h3> Exemplo 1 </h3>

Dado a lista abaixo:
```
cardapio = [['Lanche', 10.9], ['Batata',  5.5], ['Refri' ,  3.9]]
```
Acesse os valores:
1. `cardapio[0]`, `cardapio[1]` e `cardapio[2]`
2. `cardapio[0][0]`, `cardapio[0][1]`, `cardapio[1][0]`, `cardapio[1][1]`
Repare nos tipos de cada varivel

````python
# Declarando a lista de listas:
cardapio = [['Lanche', 10.9], ['Batata',  5.5], ['Refri' ,  3.9]]

# Mostra na tela o valor da lista principal: "cardapio"
print("Lista principal: {}".format(cardapio))

# Quebra de linha
print()

# Imprimindo os valores:
print("Cardapio na posio 0: {}".format(cardapio[0]))
print("Cardapio na posio 0: {}".format(cardapio[1]))
print("Cardapio na posio 0: {}".format(cardapio[2]))
print("Cardapio na posio 0, posio 0: {}".format(cardapio[0][0]))
print("Cardapio na posio 0, posio 1: {}".format(cardapio[0][1]))
````

Repare que sempre que estamos acessando os valores dentro da lista: o **primeiro** `[ ]` vai acessar as posies mais externas da lista, o **segundo** `[ ]` vai acessar as posio da lista referente  lista do **primeiro** `[ ]` e assim em diante.

Agora que j entendemos como acessar seus valores vamos ver como declarar uma lista composta.

<h2> Declarando por partes </h2>

Neste caso declara-se cada item de dentro da lista principal e depois colocam-se seus valores dentro da lista principal. Para quem est comeando  a forma mais simples de entender:

```
lanche = ['Lanche', 10.9]
batata = ['Batata',  5.5]
refri  = ['Refri' ,  3.9]
cardapio = [lanche, batata, refri]
```
No exemplo acima foram criadas 3 variveis para armazenar listas com as informaes de cada comida. A seguir essas listas foram colocadas dentro da lista principal: cardapio.

<h3> Exemplo 2 </h3>

* Declare a lista do exemplo acima.
* Imprima o valor da posio: `cardapio[1][0]`
* Compare os valores: `lanche == cardapio[0]`

````python
# Declarando a lista de listas:
lanche = ['Lanche', 10.9]
batata = ['Batata',  5.5]
refri  = ['Refri' ,  3.9]
cardapio = [lanche, batata, refri]

# Mostra a lista principal: "cardapio"
print("Lista principal: {}".format(cardapio))

# Quebra de linha
print()

# Imprimindo os valores:
print("Cardapio na posio 1, posio 0 = {}".format(cardapio[1][0]))
print("Lanche = {} e Cardpio {}".format(lanche, cardapio))
print(lanche == cardapio[0]) # Comparao de duas listas
````

<h2> Declarao com .append() </h2>


image.png

Como vimos nas aulas anteriores podemos usar o append para inserir um novo item a uma lista.

No caso de lista composta podemos usar o `.append()` para inserir uma nova lista dentro da lista mais externa.

```
cardapio = []
comida = []
comida.append('Lanche')
comida.append(10.9)
cardapio.append(comida[:])
```
Verifique que primeiro temos que declarar nossa lista externa cardapio como uma lista vazia.

A seguir criamos uma lista com os itens `comida`

Finalmente usamos o `.append()` para colocarmos esses valores dentro da lista externa: `cardapio`

<h3> Exemplo 4 </h3>

* Declare a lista cardapio inteira usando o `append()`
* Imprima o valor da posio: `cardapio[0]` e `cardapio[0][1]`

````python
# Define as listas:
cardapio = []
comida = []

# Adiciona itens a lista "comida"
comida.append("Lanche")
comida.append(10.9)

# Adiciona a lista "comida" a lista "cardapio"
cardapio.append(comida)

# Mostra a lista: "cardapio"
print("Lista Cardrdapio: {}".format(cardapio))

# Quebra de linha
print()

# Mostra na tela as posies: [0] e [0][1]
print("Posio [0] do cardpio: {}".format(cardapio[0]))
print("Posio [0][1] do cardpio: {}".format(cardapio[0][1]))
````

````python
# Declara a lista de listas:
cardapio = []
comida = []    # Varivel auxiliar

# Adiciona itens a lista "comida"
# "append" Adiciona a lista "comida" dentro da lista "cardapio"
# "clear()" limpa/exclui todos os itens da lista "comida" para ser reutilizada
comida.append("Lanche")
comida.append(10.9)
cardapio.append(comida[:]) # [:] Cria uma cpia
comida.clear()

# Adiciona itens a lista "comida"
# "append" Adiciona a lista "comida" dentro da lista "cardapio"
# "clear()" limpa/exclui todos os itens da lista "comida" para ser reutilizada
comida.append("Batata")
comida.append(3.5)
cardapio.append(comida[:]) # [:] Cria uma cpia
comida.clear()

# Adiciona itens a lista "comida"
# "append" Adiciona a lista "comida" dentro da lista "cardapio"
# "clear()" limpa/exclui todos os itens da lista "comida" para ser reutilizada
comida.append("Refri")
comida.append(5.5)
cardapio.append(comida[:]) # [:] Cria uma cpia
comida.clear()

# Mostra na tela as listas: "comida" e "cardapio"
print("Lista Cardapio: {}".format(cardapio))
print("Lista Comida: {}".format(comida))
````

Observe que tivemos que usar um mtodo novo:

```
var.clear()
```
Esse mtodo limpa a memria da varivel eliminando todos os valores que existam dentro dela.

Esse  um mtodo muito til para quando criamos algumas variveis temporrias.

## Acessando os valores de uma lista composta

Para acessar os valores de uma lista composta  s usar os `[ ]`.

Sempre  acessada a lista mais externa para a mais interna

<h2> Acesso direto </h2>

Para acessar as listas  s colocar a posio da lista desejada no primeiro `[ ]`

```
cardapio[0]    == ['Lanche', 10.9]
cardapio[1]    == ['Batata',  5.5]
cardapio[2]    == ['Refri' ,  3.9]
```
Para acessar os valores dentro das listas  s colocar a posio do valor da lista desejada no segundo `[ ]`

```
cardapio[0][1] == 10.9
cardapio[1][0] == 'Batata'
cardapio[2][0] == 'Refri'
```
