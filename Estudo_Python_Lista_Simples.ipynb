# Listas

As listas so variveis compostas mutveis.

As listas so variveis compostas que so declaradas entre colchetes `[]`

```
lista = ['var1','var2','var3']
```

Assim como as tuplas elas tambm aceitam qualquer tipo de dado.

<h2> Similaridades entre Listas e Tuplas: </h2>

Todas as funes vistas em tuplas tambm funcionam com Listas:

> Fatiamento: `[0] [0:2] [:3] [3:] [-1]` <br>
> Mtodos: `.index()`      `.count()` <br>
> Funes: `len()` e `sorted()` <br>
> Concatenar: `lista3 = lista1 + lista2` <br>
> Lista dentro de for: `for i in lista` e `for i, v in enumerate(lista):`

## Funes e Mtodos para Listas

* Modificando um termo:

```
lista[posicao] = novoValor
lista[intervalo] = novaLista
```

<h3> Exemplo 1 </h3>

Dada a lista abaixo:

`l1 = [2, 6, 4, 5, 7]`

* Troque o 3 termo por 3
* Troque os 2 primeiros termos por 0 e 1, respectivamente.

````python
# Definindo a lista
lista_Um = [2,6,4,5,7]
print(f"Lista Original: ", (lista_Um))

# Alterando um termo da lista
lista_Um[2] = 3
print(f"Lista alterada: ",(lista_Um))

# Alterando mais de um termo da lista
lista_Um[0:2] = 0, 1
print(f"Lista alterada: ", (lista_Um))

# OBS: [0:3] = 0,1 . Vai retornar como resposta uma lista menor, pois no foi informado os 3 novos termos. "0:3" so trs termos que so alterados
````

### Copiando e Clonando Listas

Quando declaramos uma nova varivel e igualamos a uma lista ela na verdade vira um clone da lista original. Qualquer modificao em uma das duas ir impactar as duas.

<h3> Exemplo 7 </h3>

Crie uma lista:
```
l1 = [0, 2, 6, 4, 8, 1, 3]
```
Em seguida faa:
```
l2 = l1
```
Delete o ltimo termo da lista `l2` e imprima as duas listas.

Observe que fazer:
```
l2 = l1
```
A gente acaba clonando as listas. Elas esto ocupando o mesmo espao de memria do computador, portanto, qualquer modificao feita em uma tambm ser feita na outra ento muito cuidado na hora de clonar as listas.

Se quisermos criar uma cpia da lista para que essa segunda possa ser modificada  vontade devemos utilizar a seguinte sintaxe:

```
l2 = l1[:]
```

<h3> Exemplo 10 </h3>

Dadas as listas abaixo:
```

```
Execute o mesmo cdigo do exemplo anterior.

Observe que o erro dado acima  porque tentamos acessar o `l2[5]` que na verdade no existe.

Para contornarmos esse erro podemos primeiro descobrir qual  a menor lista para depois colocarmos esta dentro do `enumerate()`

Mas tem um mtodo bem mais fcil que  usar a funo `zip`

```

```
Essa funo far com que o passo do for ande sempre junto para as 2 listas (ou tuplas). Assim, as variveis `val1` e `val2` vo ter sempre os valores das respectivas posies do passo do for.
