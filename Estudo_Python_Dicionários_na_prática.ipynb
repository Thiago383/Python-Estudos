## Estrutura de um Dicionrio.


image.png

A estrutura de um dicionrio  baseada em 3 pontos importantes: **keys**, **values** e **items**

Vamos observar de novo o dicionrio de exemplo:

```
         'keys'  : values
dicio = {
         'nome'  : 'Everton',       -> items
         'idade' :  29,
         'cidade': 'So Jos'
        }
```
> As `keys` so as chaves, o que substituem as posies de uma lista. Quando quisermos acessar um valor de um dicionrio  dessa chave que temos que lembrar. Assim como num Banco de Dados so as colunas que vamos acessar que precisamos saber os nomes. </br>
> Os `values` so os valores, o equivalente aos valores dentro de uma lista. So as informaes importantes do objeto que queremos acessar e fazer anlises. </br>
> Os `items` so o conjunto de `key + value`, uma novidade dos dicionrios que iro nos ajudar nos loopings.

## Dicionrios dentro de for

 comum utilizarmos os dicionrios dentro de loopings para fazermos as anlises. E para isso usaremos os 3 itens vistos acima: `values`,`keys` e `items`

<h2> Acessando os valores </h2>

image.png

```
for v in dicio.values():
  print(v)
```
Acima entende-se: *Para cada valor dentro dos valores do dicionrio, imprima valor*

<h3> Exemplo 9 </h3>

Considerando o dicionrio `pessoa` imprima cada um de seus valores

````python
# Declarando dicionrio:
pessoa = {
    "nome": "Everton",
    "idade": 29,
    "cidade": "So Jos",
    "peso": 65
}

# Mostra o contedo do Dicionrio: "pessoa"
print(pessoa)

# Quebra de linha
print()

# Imprime valores do dicionrio "pessoa"
for valor in pessoa.values():                # Para cada "valor" dentro do dicinrio "pessoa"
  print("O valor : {}".format(valor))       # Imprima na tela apenas os valores, ou seja, o resultado no pode conter as cheves: "nome, idade,..."

````

<h2> Acessando as chaves </h2>

<h2> Exerccio 1 </h2>

Dado dicionrio abaixo:
```
cardapio = {
            'lanche' : 10.9,
            'batata' : 5.5,
            'refri' : 3.9
}
```
1. Mostre o cardpio para o usurio
2. Pergunte quantos itens ele quer de cada
3. Calcule e mostre pra ele o valor total da conta

````python
# Declarando dicionrio
cardapio = {
    "lanche": 10.9,
    "batata": 5.5,
    "refri": 3.9
}

# Mostra na tela o cardpio
print("="*40)
print(f"{'Cardpio':^40}")
print("="*40)

for key, value in cardapio.items():
  print(f"{key:<32} R${value:5.2f}")
print("="*40)

# Solicita quantos itens
quantidade = {}

# Loop, pergunta quanto/quantos itens o usurio quer de cada produto
for key in cardapio.keys():
  quantidade [key] = int(input(f"Quanto de cada produto '{key}' voc quer? "))
print("-"*40)

# Calcula o valor total
soma = 0

# Loop, multiplica os valores pela quantidade e no final soma
for valor1, valor2 in zip(cardapio.values(), quantidade.values()):
  soma += valor1 * valor2

# Imprime recibo
for key, value in quantidade.items():
  print(f"{value} {key:.<20} =R${value * cardapio[key]:5.2f}")

print(f"O valor total  de: R${soma:.2f}")
````

### For dentro de For

Quando temos listas ou dicionrios aninhados,  comum precisarmos rodar for dentro de for para pegar os valores que precisamos, sendo que vai ser: 1 for para cada nvel de aninhamento

<h2> Estrutura </h2>

image.png


```
for l in listas:
  print(l)
  for k, v in l.items():
    print(k, v)
```

L-se: <br>
"*Para cada lista (dentro de listas) imprima lista*" <br>       "*Para cada item da lista selecionada imprima chave e valor*"

<h2> Exemplo 13 </h2>

Dada a lista: `pessoas` execute a estrutura acima.

````python
# Declarando lista de dicionrios:
pessoas = [
          {'nome':'Everton' , 'idade':29 , 'cidade':'So Jos'},
          {'nome':'Thiago' , 'idade':34 , 'cidade':'So Paulo'},
          {'nome':'Andr' , 'idade':22 , 'cidade':'Lorena'}
]

# montando for:
for pessoa in pessoas:
  print(pessoa)

  for key, value in pessoa.items():
    print(key, value)
````

#### Copiando dicionrios para listas


image.png

Quando queremos inserir um dicionrio dentro de uma lista, no podemos usar o mtodo do fatiamento para fazer a cpia da lista temporria, ento faremos:

```
lista.append(dicionario.copy())
```

<h2> Estrutura </h2>

```
dicio = {}
lista = []
for d in range(ini, fim):
  dicio.['Key1'] = input('Valor1 = ')
  dicio.['Key2'] = input('Valor2 = ')
  lista.append(dicio.copy())
```
Assim como nas listas, colocamos os valores inputados pelo usurio num dicionrio temporrio e a seguir o copiamos para lista.

* No podemos usar: `lista.append(dicio)` pois assim iramos clonar a lista

* No podemos usar: `lista.append(dicio[:])` pois os dicionrios no aceitam append.

<h2> Exerccio 1 </h2>

1. Crie uma lista de dicionrios contendo as notas de todos os alunos de uma sala (5 alunos). No dicionrio  para ter os campos:
* nome:
* nota1:
* nota2:
* media:
* status:

2. Pea para o usurio digitar as notas de cada aluno
3. Tire a mdia das notas e adicione no dicionrio
4. Se mdia for maior ou igual a 7, coloque "aprovado" no status, caso contrrio, coloque "reprovado"
5. Calcule a mdia das mdias e imprima para o professor.

````python
# Declarando a lista de dicionrios:
list_notas = []
dict_nota = {}


for i in range(1,6):

  dict_nota["nome"] = str(input("Digite o seu nome: "))
  dict_nota["nota1"] = float(input("Digite a nota da primeira prova: "))
  dict_nota["nota2"] = float(input("Digite a nota da segunda prova:  "))
  dict_nota["media"] = float
  dict_nota["status"] = str

  # Adicionando os "valores" (respostas do usurio) a lista: "list_notas"
  list_notas.append(dict_nota.copy())

# Mdia das notas
for pessoa in (list_notas):
  pessoa["media"] = (pessoa["nota1"] + pessoa["nota2"])/2

# Status, aprovado ou reprovado
for pessoa in (list_notas):
  if pessoa["media"] >= 7:
    pessoa["status"] = "Aprovado"
  else:
    pessoa["status"] = "Reprovado"


# Mdia das mdias
media_geral = 0
for pessoa in (list_notas):
  for key, value in pessoa.items():
    if key == "media":
      media_geral = value + media_geral

media_geral = media_geral/5
print(media_geral)
print(list_notas)

````

```
produtos = [ {
          'produto': 'maa',

         {
          'produto': 'banana',

everton = {
        'cliente': 'Everton'
        'compra': [ {
                      'produto': 'maa',

                     },
                    {

                      'qtd' = 10
                     } ]

```
1. Calcule o valor total da compra do Everton
2. Imprima a qtd que ele comprou e o que ele comprou
3. Imprima o valor total das compras

````python
# Declara Lista e Dicionrio:
produtos = [ {
          'produto': 'maa',
          'preco': 2.2
          },
         {
          'produto': 'banana',
          'preco': 1.2
          } ]

everton = {
    "cliente": "Everton",
    "compra": [{"produto": "maa", "qtd": 3}, {"produto": "banana", "qtd": 10}]
}

# Mostra a quantidade da compra
for item in everton["compra"]:
  for produto in produtos:
    if produto["produto"] == item["produto"]:
      print(f"Produto: {produto['produto']} Quantidade: {item['qtd']}")

print()


# Soma Produtos
soma_total = 0  # Varivel para armazenar o total

for item in everton["compra"]:
    for produto in produtos:
        if produto["produto"] == item["produto"]:  # Verifica se os produtos so iguais
            multiplica = produto["preco"] * item["qtd"]
            soma_total += multiplica

print(f"Total da compra: {soma_total:.2f}")
````
