# Funes
As funes so rotinas que so executadas vrias vezes nos programas.

As funes no Python e em qualquer linguagem de programao esto sendo usadas desde que voc comeou a programar. Muito provavelmente a primeira funo que voc aprendeu a usar foi a: `print()`.

 claro que voc j percebeu o tanto que essa funo  utilizada e tambm muitas outras. Mas conforme a gente vai codando mais e mais e resolvendo problemas de negcios com programao a gente percebe que est executando rotinas iguais vrias vezes e em vrios programas.

As funes vieram para dar a possibilidade para o programador criar suas prprias rotinas e assim deixar os seus cdigos mais organizados, rpidos, eficientes e ainda de quebra economizar muito tempo programando.

<h2> Estrutura da funo </h2>

```
# Funo
def funcao(parametro):     -> declarao da funcao
  print('Funo')          -> inicio da funo

                           -> fim da funo
# Programa Principal
...
funcao(parametro)                   -> chamada da funo
```
> `parametro` so as variveis, estruturas e instncias que a funo ir receber para ser executada.

<h2> Exemplo 4 </h2>

Faa um programa que tire a mdia de uma quantidade indeterminada de nmeros.

````python
# Funo de mdia
def media(*num):               # parmetro "*mum"  utilizado para acresentar quantos valores forem necessrios.
  soma = 0
  quantidade = len(num)        # "len(num)" utilizado para contar quantos parametros o progrma principal ira utilizar

  for i in num:                # Loop, Calcula a mdia
    soma += i
  media = soma / quantidade
  return media


# Principal
mostra = media(5, 8, 10, 9)
print(mostra)
````

## Escopo Local e Global

Escopo local  o escopo da funo enquanto o global  do programa inteiro.

Ao se fazer uma subrotina ou funo - ela  operada num escopo local, ou seja, todas as variveis utilizadas na subrotina NO so enviadas para o programa principal.

J as variveis declaradas no programa principal - escopo global - podem ser utilizadas nas funes mas no tem o seu valor atualizado no escopo global a no ser que use varivel global.

As variveis globais so aquelas que podem ser utilizadas tanto no escopo global quanto no local.

<h3> Exemplo 6 </h3>

Modifique o programa da soma de modo que voc imprima os valores a, b, c e s nos dois escopos.

````python
# Funo de somar
def soma(a,b=0):
  s = a + b
  a = 8
  print(a, b, c, s)
  return s

# Principal
a = 5
b = 2
c = 3
somado = soma(a, b)
print(a, b)
print(s)              # Como "s"  uma varivel local para a funo soma, ela no  visvel ou acessvel fora do escopo dessa funo, levando ao NameError.
print()
````

<h2> Declarao Global de varivel </h2>

Para contornar o problema acima  possvel declarar algumas variveis como globais. Assim qualquer modificao que elas sofrerem na subrotina tambm afetar o programa principal

```
global var
```

````python
# Funo de somar
def soma(a,b=0):
  global c, s
  s = a + b
  c = 8
  print(a, b, c, s)
  return s

# Principal
a = 5
b = 2
c = 3
print(c)             # OBS: At esse momento o "c" vale 3
somado = soma(a, b)  # Apartir dessa linha o "c" recebe uma valor fixo "c=8", pois foi definido como "global"
print(a, b, c)
print(s)
print()
````

<h2> Exerccio 1 </h2>

Faa uma funo que calcule fatorial de um nmero


````python
def fatorial():
  print("Calcula o Fatorial\n")

  # Varivel que guarda o valor inserido pelo usurio
  numero = int(input("Digite um numero inteiro: "))

  resultado = 1 # Essa varivel acumular o resultado do fatorial
  contador = 1  # Usado para iterar de 1 ate o numero inserido

  while contador <= numero:
    resultado *= contador  # "*=" ->  a mesma coisa que "a *= 3" -> "a = a * 3"
    contador += 1

  print("\nO resultado de {} : {}".format(numero, resultado))


fatorial() # Chama a funo "fatorial"
````

````python
def par_impar():

  # Interao com o usurio
  numero = int(input("Digite um nmero inteiro: "))
  resto = numero % 2

  if resto == 0:
    print("O nmero digitado  Par!!!")
  else:
    print("O nmero digitado  Impar!!!")


par_impar() # Chama a funo
````
