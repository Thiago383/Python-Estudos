# Funes
As funes so rotinas que so executadas vrias vezes nos programas.

As funes no Python e em qualquer linguagem de programao esto sendo usadas desde que voc comeou a programar. Muito provavelmente a primeira funo que voc aprendeu a usar foi a: `print()`.

 claro que voc j percebeu o tanto que essa funo  utilizada e tambm muitas outras. Mas conforme a gente vai codando mais e mais e resolvendo problemas de negcios com programao a gente percebe que est executando rotinas iguais vrias vezes e em vrios programas.

As funes vieram para dar a possibilidade para o programador criar suas prprias rotinas e assim deixar os seus cdigos mais organizados, rpidos, eficientes e ainda de quebra economizar muito tempo programando.

## Criando Funes

```

  print('Funo')

Para criar uma funo basta utilizar a palavra reservada `def` para comear a sua funo.

Todas as funes devem vir no comeo do programa (antes do programa principal) - ou em arquivo separado para ser importado. #spoiler (bibliotecas).

O que pode ser feito dentro de uma subrotina pode ser to simples quanto uma operao de soma, at complexo a nvel de um algortmo de Machine Learning.

O mais importante  sempre pensar: "eu gasto muito tempo fazendo a mesma coisa repetidas vezes?" Se sim ento crie uma subrotina! Crie uma funo!!!

<h2> Estrutura da funo </h2>

```
# Funo
def funcao(parametro):     -> declarao da funcao

<h2> Exemplo 3 </h2>

Ainda no programa da soma volte para o Programa Principal o valor da soma.

````python
# Funo de imprimir
def soma(a, b=0):
  soma = a + b
  print(f"A soma de {a} + {b} = {soma}")
  return soma


# Principal
mostra = soma(5,)   # A varivel "mostra" foi criado com o objetivo de armazenar o valor da varivel "soma".
print("Varivel 'mostra': {}".format(mostra))
````

<h2> Exemplo 4 </h2>

Faa um programa que tire a mdia de uma quantidade indeterminada de nmeros.

````python
# Funo de mdia
def media(*num):               # parmetro "*mum"  utilizado para acresentar quantos valores forem necessrios.
  soma = 0
  quantidade = len(num)        # "len(num)" utilizado para contar quantos parametros o progrma principal ira utilizar

  for i in num:                # Loop, Calcula a mdia
    soma += i
  media = soma / quantidade
  return media


# Principal
mostra = media(5, 8, 10, 9)
print(mostra)
````

## Help e Docstring

O Python possui um comando de help que nos ajuda a verificar como usar a funo corretamente.

Para usar o `help` pasta usar o comando:

```
help(funcao)
```

Aplique o help nas funes:
* print()
* len()
* sorted()

````python
help(sorted)  # Help da funo "sorted"

lista = [1, 5, 9, 6, 8, 7]
print("Utilizando a funo 'sorted' da forma convencional: {}".format(sorted(lista)))
print("")
print("Utilizando a funo 'sorted' com outros parmetros: {}".format(sorted(lista, reverse=True)))
````

````python
# Funo de mdia

def media(*num):               # parmetro "*mum"  utilizado para acresentar quantos valores forem necessrios.
  """
  A funo "media" retorna a mdia dos valores adicionados no programa "principal";
  O parmetro "(*num)" = "(num, num, num, ...)", ou seja, para acresentar quantos valores forem necessrios no programa "principal"
  Autor: Thiago de Araujo Machado
  Verso: 1.0

  """
  soma = 0
  quantidade = len(num)        # "len(num)" utilizado para contar quantos parametros o progrma principal ira utilizar

  for i in num:                # Loop, Calcula a mdia
    soma += i
  media = soma / quantidade
  return media


# Principal
mostra = media(5, 8, 10, 9)
print(mostra)
````
